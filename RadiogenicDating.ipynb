### Radiogenic dating

Consider a radioactive element $A$ decaying into a $B$, with e-folding time $\tau$. $A$ is the parent, $B$ is the daughter. Starting with concentrations $n_{A0}$ and $n_{B0}$, after a time $t$ has elapsed, the concentrations will be 

\begin{eqnarray}
n_A &=& n_{A0} e^{-t/\tau}\\
n_B &=& n_{B0} + n_{A0} (1-e^{-t/\tau})
\end{eqnarray}

The system is not solvable, as we have three unknowns ($t$, $n_{A0}$, and $n_{B0}$), and only two observables: $n_A$, and $n_B$. 

#### No initial abundance of daughter isotope

For some systems, it is possible to remove one of the two unknowns, and thus solve the system. For instance, if the daughter element is a noble gas, as in the decay of $^{40}$K into $^{40}$Ar, which occur with branching ratio $\xi=0.1$ (that is, 10% of $^{40}$K decays into $^{40}$Ar, the rest into something else).

If the rock traps the argon and there was no argon to start with, then measuring $^{40}$K and $^{40}$Ar suffices to tell the age. If the initial abundance of $^{40}$K at time $t=0$ is $n_{p0}$, then parent and daughter isotopes have abundances at time $t$


\begin{eqnarray}
n_p &=& n_{p0} e^{-t/\tau}\\
n_d &=& \xi n_{p0} (1-e^{-t/\tau})
\end{eqnarray}

Dividing, one eliminates the initial abundance 

\begin{equation}
\frac{n_p}{n_d} = \xi(e^{t/\tau}-1)
\end{equation}

which is readily solved for $t$, the age of the rock.

\begin{equation}
t=\tau \log\left(\frac{1}{\xi}\frac{n_p}{n_d}+1\right)
\end{equation}


#### Presence of initial abundance of daughter isotope

Consider a different parent-daughter pair, such as $^{87}$Rb $\rightarrow ^{87}$Sr. The rock will contain some initial strontium, so we cannot assume $n_{d0}=0$ as we did for a noble gas. In this case, we go back to the full system 

\begin{eqnarray}
n_p &=& n_{p0} e^{-t/\tau}\\
n_d &=& n_{d0} + \xi n_{p0} (1-e^{-t/\tau})
\end{eqnarray}

substituting $n_{p0}$ with the first equation

\begin{equation}
n_d = n_{d0} + \xi n_p (e^{t/\tau}-1)
\end{equation}

This is the equation of a line, of the type $y=ax+b$, where 

\begin{eqnarray}
y&=&n_d\\
x&=&n_p\\
a&=&\xi(e^{t/\tau}-1)\\
b&=&n_{d0}
\end{eqnarray}


It is usual to normalize by a reference stable isotope.

\begin{equation}
\frac{n_d}{n_{\rm ref}} = \frac{n_{d0}}{n_{\rm ref}} +  \frac{n_p}{n_{\rm ref}} \xi(e^{t/\tau}-1)
\end{equation}

Notice that $a$, the slope, is a function of age. As the rock ages, the slope increases. Samples from different rocks of same age would lie on a line.




#### Lead-lead dating

One of the most used clocks is lead-lead dating, using the decay of two Uranium isotopes with high abundance and suitable lifetimes 

\begin{eqnarray}
^{238}{\rm U} \rightarrow ^{206}{\rm Pb}   \quad (t_{1/2} = 4.47 \times 10^9 {\rm yr})\\
^{236}{\rm U} \rightarrow ^{207}{\rm Pb}  \quad (t_{1/2} = 7.04 \times 10^8 {\rm yr})
\end{eqnarray}

These systems yield the follow relations

\begin{eqnarray}
^{206}{\rm Pb} &=& ^{206}{\rm Pb}_0 +  ^{238}{\rm U}(e^{t/\tau_1}-1)\\
^{207}{\rm Pb} &=& ^{207}{\rm Pb}_0 + ^{235}{\rm U}(e^{t/\tau_2}-1)\\
\end{eqnarray}


These are equations of the type 

\begin{eqnarray}
y_1 &=& b_1 +  a_1 x_1\\
y_2 &=& b_2 + a_2x_2\\
\end{eqnarray}

with

\begin{eqnarray}
y_1&=&^{206}{\rm Pb}\\
y_2&=&^{207}{\rm Pb}\\
x_1&=&^{238}{\rm U}\\
x_2&=&^{235}{\rm U}\\
b_1&=&^{206}{\rm Pb}_0\\
b_2&=&^{207}{\rm Pb}_0\\
a_1&=&(e^{t/\tau_1}-1)\\
a_2&=&(e^{t/\tau_2}-1)
\end{eqnarray}

We can divide one by the other

\begin{equation}
\frac{y_2-b_2}{y_1-b_1} = \frac{a_2}{a_1}\frac{x_2}{x_1}
\end{equation}

if we write the RHS as $\alpha$

then 

\begin{equation}
\frac{y_2-b_2}{y_1-b_1} = \alpha
\end{equation}

where

\begin{equation}
\alpha\equiv \frac{a_2}{a_1}\frac{x_2}{x_1}
\end{equation}

isolating $y_2$

\begin{equation}
y_2 = \alpha y_1 - \alpha b_1 + b_2
\end{equation}


this is again a line, where the slope is $\alpha$ and the intercept is $\beta = b_2 -\alpha b_1$

Substituting the values, 

\begin{equation}
^{207}{\rm Pb} = \frac{^{235}{\rm U}}{^{238}{\rm U}}\left(\frac{e^{t/\tau_1}-1}{e^{t/\tau_2}-1}\right) \ ^{206}{\rm Pb} + \beta
\end{equation}


Dividing by a reference stable isotope, say, $^{204}{\rm Pb}$


\begin{equation}
\boxed{
\frac{^{207}{\rm Pb}}{^{204}{\rm Pb}} = \frac{^{235}{\rm U}}{^{238}{\rm U}}\left(\frac{e^{t/\tau_1}-1}{e^{t/\tau_2}-1}\right) \ \frac{^{206}{\rm Pb}}{^{204}{\rm Pb}} + \beta^\prime}
\end{equation}
